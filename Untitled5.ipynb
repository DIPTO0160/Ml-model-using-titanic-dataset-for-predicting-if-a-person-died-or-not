```python
df.drop(['cabin','body','boat'],axis=1,inplace=True)
```

```python
sex=pd.get_dummies(df["sex"],drop_first=True).astype(int)
```

```python
df['male']=sex
```

```python
df.drop(["male"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df["travelled_alone"].replace({"Yes":1,"No":0},inplace=True)
```

```python
df.head()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp=SimpleImputer(strategy="mean")
```

```python
df["age"]=imp.fit_transform(df[["age"]])
```

```python
df["fare"]=imp.fit_transform(df[["fare"]])
```

```python
df.isnull().sum()
```

```python
imp_frequent=SimpleImputer(strategy="most_frequent")
```

```python
df["embarked"]=imp_frequent.fit_transform(df[["embarked"]]).ravel()
```

```python
df.isnull().sum()
```

```python
 df.head()
```

```python
embark=pd.get_dummies(df["embarked"],drop_first=True).astype(int)
```

```python
 print(embark)
```

```python
df.drop(["embarked"],axis=1,inplace=True)
```

```python
df=pd.concat([df,embark],axis=1)
```

```python
df.head()
```

```python
x=df.drop(["survived"],axis=1)
y=df["survived"]
x.head()
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
mod=LogisticRegression()
mod.fit(x_train,y_train)
```

```python
pred=mod.predict(x_test)
```

```python
x_test.head()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```
